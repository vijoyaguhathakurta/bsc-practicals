# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression  # importing the function to implement linear regression
import mglearn  # library of the functions which help in machine learning 
```

```python
X,Y = mglearn.datasets.make_forge()  # generates a two-class classification dataset with 26 samples
```

```python
X,Y  # Each sample has two features and one binary target variable 
```

```python
from sklearn.model_selection import train_test_split
# importing function to split the dataset into training set and test set 
X_train, X_test, Y_train, Y_test =train_test_split(X,Y, random_state=42)
# using function to split test-train data
```

```python
X_train.shape  # size of the training set 
```

```python
X_test.shape  # size of the test set
```

```python
lrr=LogisticRegression().fit(X_train,Y_train)  # fitting the logistic regression model on training set
```

```python
lrr.score(X_train, Y_train)  # the mean accuracy of the predictions on the given training data and labels
```

```python
lrr.score(X_test,Y_test)  # the mean accuracy of the predictions on the given test data and labels
```
