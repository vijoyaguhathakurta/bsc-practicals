# K-MEANS CLUSTERING

```python
import mglearn  # library of the functions which help in machine learning 
import matplotlib.pyplot as plt  # library needed for plotting the graph
from sklearn.cluster import KMeans   # importing the class to perform K-means clustering
```

```python
mglearn.plots.plot_kmeans_algorithm()  # example of working of the algorithm
```

### For 'make_blobs' Dataset

```python
from sklearn.datasets import make_blobs  # importing the dataset make_blobs
X,Y=make_blobs(n_samples=100,random_state=1)   # assigning the data tuples and their labels to X and Y resp.
```

```python
kkmeans=KMeans(n_clusters=3)   # no. of clusters=3 
kkmeans.fit(X)  # fitting the function on the dataset
```

```python
kkmeans.predict(X)  # the cluster predicted by the algorithm for each tuple
```

```python
# plotting the clusters and the mean of each cluster 
mglearn.discrete_scatter(X[:,0],X[:,1],kkmeans.labels_,markers='o')
mglearn.discrete_scatter(kkmeans.cluster_centers_[:,0],kkmeans.cluster_centers_[:,1],[0,1,2],markers='^',markeredgewidth=2)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

### For 'make_moons' Dataset

```python
from sklearn.datasets import make_moons   # importing the dataset make_blobs
X,Y=make_moons(n_samples=250,noise=0.05,random_state=0)  # assigning the data tuples and their labels to X and Y resp.
```

```python
kkmeans=KMeans(n_clusters=2)   # no. of clusters=2 
kkmeans.fit(X)  # fitting the function on the dataset
```

```python
kkmeans.predict(X)  # the cluster predicted by the algorithm for each tuple
```

```python
# plotting the clusters and the mean of each cluster 
mglearn.discrete_scatter(X[:,0],X[:,1],kkmeans.labels_,markers='*')
mglearn.discrete_scatter(kkmeans.cluster_centers_[:,0],kkmeans.cluster_centers_[:,1],[0,1],markers='^')
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```
