```python
import numpy as np
import math
from scipy.stats import mode
```

```python
from sklearn.datasets import load_iris
iris_dataset=load_iris()
```

```python
from sklearn.metrics import accuracy_score
```

```python
from numpy.random import randint
```

```python
X=iris_dataset.data
Y=iris_dataset.target
```

```python
#test train splitting
train=list(set(randint(0,150,150)))
test=[i for i in range(150) if i not in train]
X_train=X[train]
Y_train=Y[train]
X_test=X[test]
Y_test=Y[test]
len(train)
```

```python
test_classify={}
k=11
for t in range(len(test)-1):
    A=X_test[t]
    d={}
    for i in range(len(train)-1):
        d[i]=math.dist(X_train[i],A)
    classify=[]
    for i in range(iris_dataset.target_names.shape[0]):
        classify.append(0)
    for i,s_index in zip(range(k),sorted(d,key=lambda x:d[x])):
        classify[Y_train[s_index]]=classify[Y_train[s_index]]+1
    test_classify[t]=classify.index(np.max(classify))
obs=0
for y,v in zip(Y_test,test_classify.values()):     
    if y==v:
        obs=obs+1
accuracy=1-((len(test)-obs)/len(test))
print("RESULTS:")
for data,cls in zip (X_test,test_classify.values()):
    print(data,iris_dataset.target_names[cls])
print("\nACCURACY=",accuracy)
```
