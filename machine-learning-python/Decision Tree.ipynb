# DECISION TREE CLASSIFIER on Breast Cancer Dataset 

```python
import mglearn  # library of the functions which help in machine learning 
import matplotlib.pyplot as plt  # library needed for plotting the graph
from sklearn.tree import DecisionTreeClassifier # importing the classifier
from sklearn.tree import plot_tree
```

```python
from sklearn.datasets import load_breast_cancer   # importing the dataset
cancer= load_breast_cancer()    # assigning the dataset to cancer_dataset
```

```python
from sklearn.model_selection import train_test_split # function to split the dataset into training and test set 
X_train, X_test, y_train, y_test = train_test_split(
cancer.data, cancer.target, stratify=cancer.target, random_state=42)  # using function to split test-train data
```

```python
print("Training set: ",X_train.shape,"\nTest Set: ",X_test.shape)  # size of the training and test set
```

```python
clf =DecisionTreeClassifier(max_depth=5, random_state=0)  # maximum depth of the decision tree
clf.fit(X_train, y_train)  # fitting the classifier on the training set
```

```python
clf.score(X_train, y_train) # the mean accuracy of the predictions on the given training set
```

```python
clf.score(X_test, y_test)  # the mean accuracy of the predictions on the given test set
```

```python
plt.figure(figsize=(35,8))  # generating the decision tree
plot_tree(clf,feature_names=list(cancer.feature_names),class_names=list(cancer.target_names),impurity=False,filled=True,fontsize=12)
plt.show()
```
