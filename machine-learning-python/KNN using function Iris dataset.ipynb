# K-Nearest Neighbor Classifier on the Iris Dataset (using  built-in function)

```python
import matplotlib.pyplot as plt    # library needed for plotting the graph
```

```python
from sklearn.datasets import load_iris    # importing the iris dataset 
iris_dataset=load_iris()    # assigning dataset to 'iris_dataset'
```

```python
from sklearn.metrics import accuracy_score    # importing function to find accuracy of the classifier
from sklearn.model_selection import train_test_split   # importing function to split the dataset into training set and test set 
from sklearn.neighbors import KNeighborsClassifier    # importing the KNN classifier
```

```python
X=iris_dataset.data   # assigning the data to X
Y=iris_dataset.target  #assigning the classes to Y
```

```python
X.shape
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=0)   # using function to split test-train data
```

```python
X_train.shape   # size of the training set 
```

```python
X_test.shape   # size of the test set
```

```python
training_accuracy=[]   # holds the accuracy of training set for the values of k=1 to 10
test_accuracy=[]    # holds the accuracy of test set for the values of k=1 to 10
#try KNN for K=1 to 10
k=range(1,11)
for i in k:
    clf=KNeighborsClassifier(n_neighbors=i)  # KNN classifier with k=i
    clf.fit(X_train,Y_train)   # fitting the KNN classifier on the training dataset 
    training_accuracy.append(clf.score(X_train,Y_train))  # calculating and appending the accuracy of the classifier for training set when k=i
    test_accuracy.append(clf.score(X_test,Y_test))   # calculating and appending the accuracy of the classifier for test set when k=i
```

```python
training_accuracy  # the accuracy of the classifier for training set when k=1 to 10
```

```python
test_accuracy   # the accuracy of the classifier for test set when k=1 to 10
```

```python
plt.plot(k,training_accuracy,label="Training Accuracy")   # plotting the training accuracy vs the no. of nearest neighbors
plt.plot(k,test_accuracy,label="Test Accuracy")   # plotting the test accuracy vs the no. of nearest neighbors
plt.ylabel("Accuracy")
plt.xlabel("N_Neighbors")
plt.legend()    # creating a legend to distinguish the two plots
```
