# HIERARCHICAL CLUSTERING

```python
import mglearn  # library of the functions which help in machine learning 
import matplotlib.pyplot as plt  # library needed for plotting the graph
from sklearn.cluster import AgglomerativeClustering  # importing the function to perform agglomerative clusteringimport mglearn
from scipy.cluster.hierarchy import dendrogram, ward  # importing function to create the dendrogram
```

```python
mglearn.plots.plot_agglomerative() # example of working of the algorithm
```

```python
from sklearn.datasets import make_blobs  # importing the dataset make_blobs
X,Y=make_blobs(n_samples=100,random_state=1)   # generating the dataset and assigning the data tuples and their labels to X and Y resp.
```

```python
agg=AgglomerativeClustering(n_clusters=3)  # no. of clusters=3 
assignment=agg.fit_predict(X)   # fitting the function on the dataset
```

```python
# plotting the clusters and the mean of each cluster 
mglearn.discrete_scatter(X[:,0],X[:,1],assignment,markers='^')
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

```python
X,Y=make_blobs(random_state=0,n_samples=45)  # generating the dataset and assigning the data tuples and their labels to X and Y resp.
dendrogram(ward(X))  # generate the dendogram for the dataset using ward linkage 
plt.xlabel("Sample Index")
plt.ylabel("Cluster Distance")
plt.show()   # displaying the dendogram
```
