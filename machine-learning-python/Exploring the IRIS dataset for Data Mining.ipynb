```python
import numpy as np  # numpy is a Python library used for working with arrays
```

```python
#maximum in each attribute
m=np.max(iris_dataset.data, axis=0) 
print("Maximum:\nSepal length\tSepal width\tPetal length\tPetal width")
print(m[0],"\t\t",m[1],"\t\t",m[2],"\t\t",m[3])
```

```python
#minimum in each attribute
m=np.min(iris_dataset.data, axis=0)
print("Minimum:\nSepal length\tSepal width\tPetal length\tPetal width")
print(m[0],"\t\t",m[1],"\t\t",m[2],"\t\t",m[3])
```

```python
#mean of each attribute
m=np.mean(iris_dataset.data, axis=0) 
print("Mean:\nSepal length\tSepal width\tPetal length\tPetal width")
print(round(m[0],4),"\t\t",round(m[1],4),"\t",round(m[2],4),"\t\t",round(m[3],4))
```

```python
#median of each attribute
m=np.median(iris_dataset.data, axis=0) 
print("Median:\nSepal length\tSepal width\tPetal length\tPetal width")
print(m[0],"\t\t",m[1],"\t\t",m[2],"\t\t",m[3])
```

```python
#mode of each attribute
from scipy import stats
m,c=stats.mode(iris_dataset.data, axis=0, keepdims=True)
print("Median:\nSepal length\tSepal width\tPetal length\tPetal width")
print(m[0][0],"\t\t",m[0][1],"\t\t",m[0][2],"\t\t",m[0][3])
```

```python
#standard deviation of each attribute
m=np.std(iris_dataset.data, axis=0)
print("Standard Deviation:\nSepal length\tSepal width\tPetal length\tPetal width")
print(round(m[0],4),"\t\t",round(m[1],4),"\t",round(m[2],4),"\t",round(m[3],4))
```

```python
#variance of each attribute
m=np.var(iris_dataset.data, axis=0) 
print("Variance:\nSepal length\tSepal width\tPetal length\tPetal width")
print(round(m[0],4),"\t\t",round(m[1],4),"\t",round(m[2],4),"\t",round(m[3],4))
```

```python
X=iris_dataset.data
Y=iris_dataset.target
```

```python
#plotting the graphs showing the clusters and relation between the attributes
iris_dataframe=pd.DataFrame(X,columns=iris_dataset.feature_names)
grr=pd.plotting.scatter_matrix(iris_dataframe,c=Y,figsize=(15,15),marker='o',hist_kwds={'bins':20},s=60,alpha=.8,cmap=mglearn.cm3)
```
