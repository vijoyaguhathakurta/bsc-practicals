# LINEAR REGRESSION 

```python
from sklearn.linear_model import LinearRegression  # importing the function to implement linear regression 
import mglearn   # library of the functions which help in machine learning 
```

```python
X,Y=mglearn.datasets.make_wave(n_samples=70)  # generates a one-dimensional regression dataset
X.shape
```

```python
from sklearn.model_selection import train_test_split
# importing function to split the dataset into training set and test set 
X_train, X_test, Y_train, Y_test =train_test_split(X,Y, random_state=42)
# using function to split test-train data
```

```python
X_train.shape  # size of the training set 
```

```python
X_test.shape  # size of the test set
```

```python
lr=LinearRegression().fit(X_train,Y_train)   # fitting the linear regression model on training set
```

```python
lr.coef_  # array containing the coefficients of the features in the decision function of regression model
```

```python
lr.intercept_  # array containing the intercept added to the decision function of the regression model 
```

```python
lr.score(X_train, Y_train)  # the mean accuracy of the predictions on the given training data and labels
```

```python
lr.score(X_test,Y_test)  # the mean accuracy of the predictions on the given test data and labels
```

```python
"""generates a scatter plot of the dataset with the feature on the x-axis and the target variable on the y-axis 
and shows the line of best fit computed by the linear regression model."""
mglearn.plots.plot_linear_regression_wave()  
```
