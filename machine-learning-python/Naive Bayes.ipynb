# Predicting whether to play or not based on weather using NAIVE-BAYES CLASSIFIER

```python
import pandas as pd  #  importing library used for faster data analysis and data pre-processing.
import matplotlib.pyplot as plt  # for plotting the graph
from sklearn.metrics import confusion_matrix  # function to generate confusion matrix
from sklearn.naive_bayes import GaussianNB  #  importing the classifier
```

```python
data=pd.read_csv("Weather.csv")
data
```

```python
from sklearn.preprocessing import LabelEncoder  # importing the class used to encode target labels with values between 0 and n_classes-1
le= LabelEncoder()  # creating an instance of the LabelEncoder class
 # fitting the encoder to transform the target labels
data['Outlook']=le.fit_transform(data['Outlook'])
data['Temp']=le.fit_transform(data['Temp'])
data['Humidity']=le.fit_transform(data['Humidity'])
data['Windy']=le.fit_transform(data['Windy'])
data['Play']=le.fit_transform(data['Play'])
target=data['Play'] # seperating the target from data
data=data.drop('Play',axis='columns')
print(target)  # labels
data # data tuples
```

```python
clf=GaussianNB()
clf.fit(data,target)
Y_pred=clf.predict(data)
clf.score(data,target)  # accuracy of the classifier
```

```python
cm = confusion_matrix(target, Y_pred) # Generate the confusion matrix
# Plot the confusion matrix
plt.matshow(cm, cmap='Greys')
plt.title('Confusion Matrix')
plt.colorbar()
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```
